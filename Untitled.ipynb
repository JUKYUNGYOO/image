```
# Colab은 Jupyter notebook을 기반으로 동작하므로 Matplotlib을 이용하여 
# 이미지를 출력합니다.
# opencv는 BGR을 기준으로 하며, Matplotlib는 RGB를 기준으로  합니다 .

import cv2
import matplotlib.pyplot as plt

img_basic = cv2.imread('cat.jpg',cv2.IMREAD_COLOR)
plt.imshow(cv2.cvtColor(img_basic,cv2.COLOR_BGR2RGB))
# opencv는 BGR로 읽는데, matplotlib는 RGB로 표현하므로
# BGR2RGB해주기 
plt.show()

img_basic = cv2.cvtColor(img_basic,cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(img_basic,cv2.COLOR_GRAY2RGB))
plt.show()

img_basic = cv2.imread('cat.jpg',cv2.IMREAD_COLOR)
cv2.waitKey(0)
# 이미지가 보여진 후 바로 꺼지지 않도록
cv2.imwrite('result1.png',img_basic)

```

```
import cv2
import matplotlib.pyplot as plt
import time

image = cv2.imread('cat.jpg')

start_time = time.time()
for i in range(0,100):
  for j in range(0,100):
    image[i,j] = [255,255,255]
print("---%s seconds ---" % (time.time()-start_time))


start_time = time.time()
image[0:100,0:100] = [0,0,0]
print("---%s seconds---" % (time.time()-start_time))

plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB)) 
plt.show()
```

```
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('cat.jpg')

# Numpy Slicing : ROI 처리 가능
roi = image[200:350,50:200]

# ROI단위로 이미지 복사하기 
image[0:150,0:150] = roi

plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.show()
```

```
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('cat.jpg')
image[:,:,2] = 0
# BGR 픽셀값에 R에 해당되는 부분을 0으로 만드니깐 
# 모든 색에 대해서 Red 부분이 사라짐.
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.show()
```
